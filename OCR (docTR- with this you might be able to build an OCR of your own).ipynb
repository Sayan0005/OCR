```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Load the pre-trained OCR model
model = ocr_predictor(pretrained=True)

# Load the image
image_path = "/content/normal image.jpg"  # Replace with the path to your image
doc = DocumentFile.from_images(image_path)

# Perform OCR
result = model(doc)

# Extract text
text = result.export()  # Extracts as a nested dictionary
print("Extracted Text:")
for page in result.pages:
    print(page.content)  # Prints the text from each page

# Optional: Save results to JSON
import json
with open("ocr_result.json", "w") as f:
    json.dump(text, f)
    print("OCR results saved to ocr_result.json")
```

```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Load the OCR model
model = ocr_predictor(pretrained=True)

# Load the image
image_path = "/content/normal image.jpg"
doc = DocumentFile.from_images(image_path)

# Analyze the document
result = model(doc)

# Extract blocks and handle text safely
for page in result.pages:
    blocks = page.export().get("blocks", [])
    for block in blocks:
        # Safely access the "text" key
        if "text" in block:
            print(block["text"])
        else:
            print("Block without text key:", block)

# Optional: Save the result to JSON
json_output = result.export()
with open("ocr_results.json", "w") as f:
    import json
    json.dump(json_output, f, indent=4)
```

```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Load the OCR model
model = ocr_predictor(pretrained=True)

# Load the image
doc = DocumentFile.from_images("/content/normal image.jpg")

# Analyze the document using the model
result = model(doc)

# Extract and display the text and confidence scores
for page_idx, page in enumerate(result.pages):
    print(f"Page {page_idx + 1}:")
    for block in page.blocks:
        for line in block.lines:
            for word in line.words:
                text = word.value
                confidence = word.confidence
                print(f"Text: {text}, Confidence: {confidence:.2f}")
```

```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Load the OCR model
model = ocr_predictor(pretrained=True)

# Load the image
doc = DocumentFile.from_images("/content/normal image.jpg")

# Analyze the document using the model
result = model(doc)

# Retrieve text in its spatial format
formatted_text = ""
for page_idx, page in enumerate(result.pages):
    formatted_text += f"Page {page_idx + 1}:\n"
    for block in page.blocks:
        for line in block.lines:
            line_text = " ".join(word.value for word in line.words)
            formatted_text += f"{line_text}\n"  # Preserve line structure
    formatted_text += "\n"  # Space between pages

# Save or print the formatted text
print(formatted_text)

# Optionally save to a file
with open("formatted_text.txt", "w") as f:
    f.write(formatted_text)
```

```python
import os

# Set the OpenAI API key
os.environ["OPENAI_API_KEY"] = api_key
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
from pydantic import BaseModel, Field
from typing import Optional
from datetime import date

class InvoiceDetails(BaseModel):
    supplier_name: Optional[str] = Field(None, description="Name of the supplier")
    supplier_address_line1: Optional[str] = Field(None, description="First line of the supplier's address")
    supplier_city: Optional[str] = Field(None, description="City where the supplier is located")
    supplier_state: Optional[str] = Field(None, description="State where the supplier is located")
    supplier_zip: Optional[str] = Field(None, description="ZIP code of the supplier's location")
    invoice_number: Optional[str] = Field(None, description="The invoice number")
    invoice_delivery_date: Optional[date] = Field(None, description="The date the invoice was delivered (YYYY-MM-DD)")
    customer_account_number: Optional[str] = Field(None, description="Customer's account number")
    sold_bill_to_restaurant_name: Optional[str] = Field(None, description="Name of the sold bill-to restaurant")
    sold_bill_to_restaurant_address_line1: Optional[str] = Field(None, description="First line of the sold bill-to restaurant's address")
    sold_bill_to_restaurant_city: Optional[str] = Field(None, description="City of the sold bill-to restaurant")
    sold_bill_to_restaurant_state: Optional[str] = Field(None, description="State of the sold bill-to restaurant")
    sold_bill_to_restaurant_zip: Optional[str] = Field(None, description="ZIP code of the sold bill-to restaurant")
    ship_to_restaurant_name: Optional[str] = Field(None, description="Name of the ship-to restaurant")
    ship_to_restaurant_address_line1: Optional[str] = Field(None, description="First line of the ship-to restaurant's address")
    ship_to_restaurant_city: Optional[str] = Field(None, description="City of the ship-to restaurant")
    ship_to_restaurant_state: Optional[str] = Field(None, description="State of the ship-to restaurant")
    ship_to_restaurant_zip: Optional[str] = Field(None, description="ZIP code of the ship-to restaurant")
    sales_total_gross: Optional[float] = Field(None, description="Total gross sales")
    charges: Optional[float] = Field(None, description="Charges applied to the order")
    order_total: Optional[float] = Field(None, description="Total order amount")
    subtotal: Optional[float] = Field(None, description="Subtotal amount before charges or taxes")

model_with_tools = model.bind_tools([InvoiceDetails])
```
